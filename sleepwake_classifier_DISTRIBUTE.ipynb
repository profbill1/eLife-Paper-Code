```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Read in and clean data

```python
# Read and clean up data from spreadsheet
DATA = pd.read_excel('sleep_wave_metrics.xlsx')

# drop rows with missing data
DATA = DATA.dropna()
```

```python
# define data and target
def bin_developmental_stages(row):
    if row['Stage'] == 'P1' or row['Stage'] == 'P2':
        val = 1
    elif row['Stage'] == 'P3' or row['Stage'] == 'P5':
        val = 1
    else:
        val = 2
    return val

def developmental_day(row):
    Pval = row['Stage']
    return int(Pval[1])

DATA['Target'] = DATA.apply(bin_developmental_stages, axis=1)
DATA['PStage'] = DATA.apply(developmental_day, axis=1)    
DATA    
```

```python
# colormap for coloring developmental days
from matplotlib.colors import LinearSegmentedColormap

CubicL = [(0.4706,         0,    0.5216),
             (0.5137,    0.0527,    0.7096),
             (0.4942,    0.2507,    0.8781),
             (0.4296,    0.3858,    0.9922),
             (0.3691,    0.5172,    0.9495),
             (0.2963,    0.6191,    0.8515),
             (0.2199,    0.7134,    0.7225),
             (0.2643,    0.7836,    0.5756),
             (0.3094,    0.8388,    0.4248),
             (0.3623,    0.8917,    0.2858),
             (0.5200,    0.9210,    0.3137),
             (0.6800,    0.9255,    0.3386),
             (0.8000,    0.9255,    0.3529),
             (0.8706,    0.8549,    0.3608),
             (0.9514,    0.7466,    0.3686),
             (0.9765,    0.5887,    0.3569)]

cm = LinearSegmentedColormap.from_list(
        '', CubicL, N=8)
```

```python
X = DATA.loc[:, 'SleepDur':'SleepWaveFreq']


h = plt.scatter(X.loc[:,'Motor'], X.loc[:,'SleepWaveFreq'], 
            s=60, c=DATA.PStage, cmap=cm)

cbar = plt.colorbar(h)
cbar.set_ticks(np.arange(1.3,9,1.75))
cbar.ax.set_yticklabels('')
cbar.ax.set_yticklabels(['P1', 'P3', 'P5', 'P7'])
```

## Train classifier, report cross-validated accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# regularize the data
X = DATA.loc[:, 'SleepDur':'SleepWaveFreq']
X = StandardScaler().fit_transform(X)


# linear SVC model
myclf = SVC(kernel='linear')

# now do that many, many times
reps = 1000
scores = np.zeros(reps)
coefs = np.zeros([reps, X.shape[1]])

for it in range(reps):
    # test/train split
    X_train, X_test, y_train, y_test = train_test_split(X,
                         DATA.Target, 
                         test_size=.2)

    # fit on training data
    myclf.fit(X_train, y_train)
    
    coefs[it] = myclf.coef_
    scores[it] = myclf.score(X_test, y_test)
    
print('Mean cross-validated accuracy: ', scores.mean())
print('Std of cv accuracy', scores.std())
```

## Unsupervised analyses

```python
# let's start with a simple SVD
U, Sig, V = np.linalg.svd(X)
plt.plot(Sig)

plt.xlabel('Number of singular values')
plt.ylabel('Magnitude of singular value')
```

```python
plt.scatter(U[:, 0], U[:, 1], s=40, c=DATA.PStage, cmap=cm)

plt.xlabel('PC1')
plt.xlabel('PC2')
```

The data does not make too much sense projected on the first 2 PC's, so let's try something more nonlinear instead.

```python
from sklearn import manifold

X = DATA.loc[:, 'SleepDur':'SleepWaveFreq']

from matplotlib.pyplot import figure
figure(num=None, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')


neighbors = np.arange(2, 10)

for K in neighbors:
    se = manifold.SpectralEmbedding(n_components=2, n_neighbors=K)

    Y = se.fit_transform(X)

    plt.subplot(3, 3, K+1-neighbors.min())
    h = plt.scatter(Y[:,0], Y[:,1], s=40, c=DATA.PStage, cmap=cm)
    
plt.subplot(3, 3, 9)    
plt.colorbar(h)
```

```python
figure(num=None, figsize=(5, 4), dpi=300, facecolor='w', edgecolor='k')

X = DATA.loc[:, 'SleepDur':'SleepWaveFreq']

se = manifold.SpectralEmbedding(n_components=2, n_neighbors=5)
Y = se.fit_transform(X)

ah = plt.figure()
h = plt.scatter(Y[:,0], Y[:,1], s=70, 
                c=DATA.PStage, 
                cmap=cm,
                edgecolors='k')

frame1 = plt.gca()
# frame1.axes.xaxis.set_ticklabels([])
# frame1.axes.yaxis.set_ticklabels([])

cbar = plt.colorbar(h)
cbar.set_ticks(np.arange(1.3,9,1.75))
cbar.ax.set_yticklabels('')
cbar.ax.set_yticklabels(['P1', 'P3', 'P5', 'P7'])

plt.xlabel('SE1')
plt.ylabel('SE2')

plt.savefig('SpectralEmbedding_by_Age.eps')
```
